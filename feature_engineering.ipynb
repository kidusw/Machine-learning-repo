```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def createSample(mu=50,st=13,sample_size=100,seed=42):
    np.random.seed(seed)
    sample_ages=np.random.normal(loc=mu,scale=13,size=sample_size)
    sample_ages=np.round(sample_ages,decimals=0)
    
    return sample_ages
```

```python
sample=createSample()
```

```python
100*new_df.isnull().sum()/len(new_df)
```

```python
def percentage(df):
    percentage_nan=100*df.isnull().sum()/len(df)
    percentage_nan=percentage_nan[percentage_nan>0].sort_values()
    
    return percentage_nan
```

```python
percent_nan=percentage(new_df)
```

```python
percent_nan
```

```python
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.xticks(rotation=90)
plt.ylim(0,1)
```
