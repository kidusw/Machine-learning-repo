```python
X=df.drop('species',axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler=StandardScaler()
```

```python
scaled_x_train=scaler.fit_transform(X_train)
scaled_x_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression(solver='saga',multi_class='ovr',max_iter=5000)
```

```python
penalty=['l1','l2','elasticnet']
l1_ratio=np.linspace(0,1,10)
C=np.logspace(0,5,10)
param_grid={
    "penalty":penalty,
    "l1_ratio":l1_ratio,
    "C":C,
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model=GridSearchCV(estimator=log_model,param_grid=param_grid)
```

```python
grid_model.fit(scaled_x_train,y_train)
```

```python
grid_model.best_params_
```

```python
y_pred=grid_model.predict(scaled_x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score,confusion_matrix,ConfusionMatrixDisplay
```

```python
accuracy_score(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred,normalize='pred')
```
